```python
import cv2
import numpy as np

import matplotlib.pyplot as plot
%matplotlib inline
```

```python
# 3:color - dtype: data type
blankImg = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
plot.imshow(blankImg)
```

```python
# pt1: left-upper point - pt2: right-bottom point
# color: RGB
# thickness
# It override the image, so it's no longer a blank image
cv2.rectangle(blankImg, pt1=(206,206), pt2=(306,306), color=(0,0,255), thickness=10)
```

```python
plot.imshow(blankImg)
```

```python
cv2.circle(img=blankImg, center=(100, 100), radius=50, color=(255,0,0), thickness=8)
plot.imshow(blankImg)
```

```python
# thickness=-1 draws a filled in figure.
cv2.circle(img=blankImg, center=(400, 400), radius=40, color=(200,255,0), thickness=-1)
plot.imshow(blankImg)
```

```python
cv2.line(img=blankImg, pt1=(50,400), pt2=(300, 450), color=(50,50,50), thickness=5)
plot.imshow(blankImg)
```

```python
img = cv2.imread('DATA/00-puppy.jpg')
cv2.line(img,(0,0), (img.shape[1],img.shape[0]), (255,255,255), 10)
cv2.line(img,(img.shape[1],0), (0,img.shape[0]), (255,255,255), 10)
plot.imshow(img)
```
